```python
import cv2
import matplotlib.pyplot as plt
```

```python


# Load the images
elephant_image_path = './elephant2.png'
mask_image_path = './segmented_mask.png'
```

```python


# Read the images
elephant_image = cv2.imread(elephant_image_path)
mask_image = cv2.imread(mask_image_path, cv2.IMREAD_GRAYSCALE)

# Resize the mask to match the dimensions of the elephant image
mask_image = cv2.resize(mask_image, (elephant_image.shape[1], elephant_image.shape[0]))

# Apply the mask using bitwise_and
masked_image = cv2.bitwise_and(elephant_image, elephant_image, mask=mask_image)

# Convert BGR to RGB for displaying correctly with matplotlib
masked_image_rgb = cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB)

# Display the result
plt.imshow(masked_image_rgb)
plt.axis('off')  # Hide the axes
plt.show()
```
