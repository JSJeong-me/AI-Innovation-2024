```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 검은색 배경에 흰색 사각형 생성
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -1)
```

```python
# 검은색 배경에 흰색 원 생성
circle = np.zeros((300, 300), np.uint8)
cv2.circle(circle, (150, 150), 125, 255, -1)
```

```python
# bitwise AND
bitwise_and = cv2.bitwise_and(square, circle)
```

```python
# bitwise OR
bitwise_or = cv2.bitwise_or(square, circle)

# bitwise XOR
bitwise_xor = cv2.bitwise_xor(square, circle)

# bitwise NOT (square에 대한 NOT 연산)
bitwise_not = cv2.bitwise_not(square)
```

```python
# 결과 출력 함수
def show_image(title, image):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# 원본 이미지와 bitwise 연산 결과 출력
show_image("Square", square)
show_image("Circle", circle)
show_image("Bitwise AND", bitwise_and)
show_image("Bitwise OR", bitwise_or)
show_image("Bitwise XOR", bitwise_xor)
show_image("Bitwise NOT (Square)", bitwise_not)
```
