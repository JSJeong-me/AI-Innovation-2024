```python
import cv2
import numpy as np

# Load the images
banana_image_path = './banana3.jpg'
banana_mask_path = './banana-mask.png'
```

```python

# Read the original image and mask
banana_image = cv2.imread(banana_image_path)
banana_mask = cv2.imread(banana_mask_path, cv2.IMREAD_GRAYSCALE)

# Check if resizing is needed by comparing their dimensions
banana_image_shape = banana_image.shape
banana_mask_shape = banana_mask.shape

banana_image_shape, banana_mask_shape
```

```python
# Resize the mask to match the banana image dimensions
resized_banana_mask = cv2.resize(banana_mask, (banana_image_shape[1], banana_image_shape[0]))
```

```python
resized_banana_mask.shape
```

```python
# Apply the resized mask using bitwise_and
masked_banana_image = cv2.bitwise_and(banana_image, banana_image, mask=resized_banana_mask)
```

```python
# Save the result
masked_banana_output_path = './masked_banana.jpg'
cv2.imwrite(masked_banana_output_path, masked_banana_image)

masked_banana_output_path
```
