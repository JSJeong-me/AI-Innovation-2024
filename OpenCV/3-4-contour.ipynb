```python
import sys
import random
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1ERYkBsk0uqDm4QdwOO8-OMnfuLXYevGX' -O coins.png
```

```python
src = cv2.imread('coins.png')

if src is None:
    print('Image load failed!')
    sys.exit()
```

```python
cv2_imshow(src)
```

```python
# src = cv2.resize(src, (0, 0), fx=0.5, fy=0.5)
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(src_gray)
```

```python
h, w = src.shape[:2]
dst1 = np.zeros((h, w, 3), np.uint8)
dst2 = np.zeros((h, w, 3), np.uint8)
```

```python
cv2_imshow(dst2)
```

```python
# 이진화
_, src_bin = cv2.threshold(src_gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2_imshow(src_bin)
```

```python
# 외곽선 검출
contours1, _ = cv2.findContours(src_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contours2, _ = cv2.findContours(src_bin, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for i in range(len(contours1)):
#    c = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    cv2.drawContours(dst1, contours1, i, (0,255,0), 1)

for i in range(len(contours2)):
#    c = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    cv2.drawContours(dst2, contours2, i, (0,255,0), 1)

# cv2.imshow('src', src)
# cv2.imshow('src_bin', src_bin)
# cv2.imshow('dst1', dst1)
# cv2.imshow('dst2', dst2)

cv2_imshow(dst1)
cv2_imshow(dst2)

# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
