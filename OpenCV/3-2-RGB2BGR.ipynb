```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1nSWbWKOA-NN0otRR_-8rgFKUw-hZuuP1' -O giraffe.jpg
```

```python
src = cv2.imread('./giraffe.jpg', cv2.IMREAD_COLOR) # cv2.IMREAD_COLOR, cv2.IMREAD_GRAYSCALE, cv2.IMREAD_UNCHANGED
```

```python
cv2_imshow(src)
```

```python
src = cv2.imread('./giraffe.jpg')
```

```python
b,g,r = cv2.split(src)
```

```python
reverse_img = cv2.merge((r,g,b))
```

```python
cv2_imshow(reverse_img)
```

```python
import matplotlib.pyplot as plt
```

```python
#BGR
```

```python
# BGR to RGB
src_rgb = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(src_rgb)
plt.show()
```
