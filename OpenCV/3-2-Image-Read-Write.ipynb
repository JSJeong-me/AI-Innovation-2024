# 1. Import Dependencies

```python
from google.colab.patches import cv2_imshow
import cv2
import os
```

# 2. Read in Image

```python
img_path = os.path.join('elephant1.jpg')
```

```python
img = cv2.imread(img_path)
```

```python
cv2_imshow(img)
```

# 3. Render Image

```python
# !pip install matplotlib
```

```python
from matplotlib import pyplot as plt
```

```python
# Show image
plt.imshow(img)
```

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(recolor)
# plt.imshow(gray)
plt.show()
```

# 4. Write out Image

```python
cv2.imwrite('grayimage.jpg', gray)
```

```python
output_path = os.path.join('outputgray.jpg')
```
