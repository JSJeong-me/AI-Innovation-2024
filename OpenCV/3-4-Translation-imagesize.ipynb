```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지 불러오기 (사용자가 원하는 이미지를 여기에 넣으세요)
img = cv2.imread('elephant2.png')

# 이미지를 BGR에서 RGB로 변환 (OpenCV는 BGR 형식이므로 변환 필요)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Translation 행렬 정의 (x축으로 50px, y축으로 100px 이동)
tx, ty = 50, 100
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# 이미지 크기 가져오기
height, width = img.shape[:2]

# Translation 적용
translated_img = cv2.warpAffine(img, translation_matrix, (width, height))

# 결과 시각화
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
translated_img_rgb = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)
plt.imshow(translated_img_rgb)
plt.title('Translated Image')

plt.show()
```
