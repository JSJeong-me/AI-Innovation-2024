```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image_path = './banana3.jpg'
image = cv2.imread(image_path)
```

```python


# Convert to RGB (OpenCV loads as BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Use a threshold to segment the image
_, threshold_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw bounding boxes around detected objects (bananas)
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (255, 0, 0), 3)

# Display the result with bounding boxes
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
# plt.imshow(gray_image, cmap='gray')
# plt.title('Bananas Detected')
plt.axis('off')  # Turn off axis
plt.show()
```
