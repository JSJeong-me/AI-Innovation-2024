# 4. Working with Video Captures

```python
# Establish capture
cap = cv2.VideoCapture(os.path.join('20201009.mp4'))
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):

    # Read frame
    ret, frame = cap.read()

    # Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Show image
    cv2_imshow(gray)

    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

#Close down everything
cap.release()
# cv2.destroyAllWindows()
```

# 5. Writing out Videos

```python
# Establish capture
cap = cv2.VideoCapture(os.path.join('20201009.mp4'))

# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
# Video Writer
video_writer = cv2.VideoWriter(os.path.join('output.avi'), cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'), fps, (width, height), isColor=False)
# cv2.VideoWriter_fourcc('P','I','M','1'): The VideoWriter_fourcc function specifies the codec used for compressing the video. 'P','I','M','1' refers to MPEG-1 codec
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):

    # Read frame
    ret, frame = cap.read()

    # Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Show image
    cv2_imshow(gray) #cv2.imshow('Video Player', gray)

    # Write out frame
    video_writer.write(gray)

    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
# Release video writer
video_writer.release()
```
