```python


# 임베딩 추출을 위한 라이브러리 로드
from transformers import AutoTokenizer, AutoModel
import torch

# 사용할 모델과 토크나이저 로드 (여기서는 'sentence-transformers/all-MiniLM-L6-v2' 사용)
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python


# 임베딩 추출을 위한 함수 정의
def get_sentence_embedding(sentence):
    # 입력 문장을 토큰화
    inputs = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True)

    # 모델을 통해 임베딩 추출
    with torch.no_grad():
        outputs = model(**inputs)

    # 임베딩은 모델의 마지막 은닉층 출력 중 [CLS] 토큰의 표현을 사용
    sentence_embedding = outputs.last_hidden_state[:, 0, :]
    return sentence_embedding
```

```python


# 예제 문장
sentence = "문장 임베딩을 생성해봅시다."
```

```python


# 임베딩 생성
embedding = get_sentence_embedding(sentence)
print(embedding)
```

## 문장간의 유사도 확인

```python


# 임베딩 추출을 위한 라이브러리 로드
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

# 사용할 모델과 토크나이저 로드 (여기서는 'sentence-transformers/all-MiniLM-L6-v2' 사용)
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# 임베딩 추출을 위한 함수 정의
def get_sentence_embedding(sentence):
    # 입력 문장을 토큰화
    inputs = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True)

    # 모델을 통해 임베딩 추출
    with torch.no_grad():
        outputs = model(**inputs)

    # 임베딩은 모델의 마지막 은닉층 출력 중 [CLS] 토큰의 표현을 사용
    sentence_embedding = outputs.last_hidden_state[:, 0, :]
    return sentence_embedding

# 코사인 유사도 계산 함수
def cosine_similarity(embedding1, embedding2):
    return F.cosine_similarity(embedding1, embedding2).item()

# 예제 문장
sentence1 = "문장 임베딩을 생성해봅시다."
sentence2 = "우리는 문장 간의 유사도를 계산할 것입니다."

# 임베딩 생성
embedding1 = get_sentence_embedding(sentence1)
embedding2 = get_sentence_embedding(sentence2)

# 문장 간 유사도 계산
similarity = cosine_similarity(embedding1, embedding2)
print(f"문장 간 유사도: {similarity:.4f}")
```
