```python
# 필요한 라이브러리 설치
!pip install transformers

# 토크나이저 불러오기
from transformers import AutoTokenizer

# 토크나이저를 얻는 함수
def get_tokenizer(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    return tokenizer

# 예제: BERT 모델의 토크나이저 로드
model_name = 'bert-base-uncased'
tokenizer = get_tokenizer(model_name)

# 토크나이저 사용 예시
text = "Hello, how are you?"
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")
```

```python

# torchtext에서 필요한 모듈 가져오기
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator

# 토크나이저를 얻는 함수
def get_tokenizer_example(tokenizer_name='basic_english'):
    tokenizer = get_tokenizer(tokenizer_name)
    return tokenizer

# 토크나이저 사용 예시
tokenizer = get_tokenizer_example()

# 예제 텍스트
text = "Hello, how are you?"

# 텍스트를 토큰화
tokens = tokenizer(text)

# 토큰을 이용하여 Vocabulary 생성
def yield_tokens(data):
    for text in data:
        yield tokenizer(text)

# 예제 데이터로 Vocabulary 생성
vocab = build_vocab_from_iterator(yield_tokens([text]), specials=["<unk>"])
vocab.set_default_index(vocab["<unk>"])

# 토큰을 ID로 변환
token_ids = [vocab[token] for token in tokens]

print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")
```
