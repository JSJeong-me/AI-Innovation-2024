```python
from PIL import Image
import torchvision.transforms as transforms
import torch

# Load the image
image_path = 'elephant1.jpg'
image = Image.open(image_path)

# Define a transform to convert the image to a tensor
transform = transforms.Compose([
    transforms.ToTensor()  # Converts [0, 255] range to [0.0, 1.0] range
])

# Apply the transform to the image
image_tensor = transform(image)

# Display the tensor shape
print(f"Image tensor shape: {image_tensor.shape}")
```

```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms

# Assuming 'image_tensor' is the tensor we created
# Convert the tensor back to a PIL Image
transform_to_pil = transforms.ToPILImage()
image = transform_to_pil(image_tensor)

# Display the image
plt.imshow(image)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
import matplotlib.pyplot as plt
import torch

# Assuming image_tensor is the tensor we previously loaded
# image_tensor shape: torch.Size([3, 800, 1200])

# Extract the individual channels
red_channel = image_tensor[0, :, :].numpy()  # Red channel
green_channel = image_tensor[1, :, :].numpy()  # Green channel
blue_channel = image_tensor[2, :, :].numpy()  # Blue channel

# Plot each channel as a grayscale image
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms as transforms
import torch

# Load the image
image_path = './elephant2.png'
image = Image.open(image_path)

# Convert the image to RGBA to add a fourth channel
image_rgba = image.convert("RGBA")

# Define a transform to convert the RGBA image to a tensor
transform = transforms.Compose([
    transforms.ToTensor()  # Converts [0, 255] range to [0.0, 1.0] range
])

# Apply the transform to the image
image_tensor = transform(image_rgba)

# Extract the individual layers
red_channel = image_tensor[0, :, :].numpy()    # Red channel
green_channel = image_tensor[1, :, :].numpy()  # Green channel
blue_channel = image_tensor[2, :, :].numpy()   # Blue channel
alpha_channel = image_tensor[3, :, :].numpy()  # Alpha channel

# Plot each channel as an image
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

axes[3].imshow(alpha_channel, cmap='gray')
axes[3].set_title('Alpha/Grayscale Channel')
axes[3].axis('off')

plt.show()
```
